```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from jupyterthemes import jtplot
jtplot.style()
```

```python
import PIL
from PIL import Image
im = Image.open("bird_small.png")
display(im) 
```

```python
im_arr = np.array(im)
im_arr = im_arr/255 #don't forget to do backward
X = im_arr.reshape(-1, 3)
print(X.shape)
```

```python
def find_closest_centroids(X, centroids):
    idx = []
    for i in range(X.shape[0]):
        idx.append(np.argmin(np.sum((X[i] - centroids)**2, axis=1)/X.shape[1]))
    return np.array(idx)
```

```python
def compute_centroids(X, idx, K):
    centroids = np.zeros([K, X.shape[1]])
    for i in range(K):
        X_i = X[idx == i]
        centroids[i] = np.sum(X_i, axis=0)/X_i.shape[0]
    return centroids
```

```python
def run_kMeans(X, centroids, max_iters):
    centroids_history = [centroids]
    for i in range(max_iters):
        idx = find_closest_centroids(X, centroids)
        centroids = compute_centroids(X, idx, K)
        centroids_history.append(centroids)
    return idx, centroids, np.array(centroids_history)
```

```python
def kMeans_init_centroids(X, K):
    return X[np.random.choice(X.shape[0], K, replace=False), :]
```

```python
K = 16
max_iters = 10
initial_centroids = kMeans_init_centroids(X, K)
```

```python
idx, res_centroids, history = run_kMeans(X, initial_centroids, max_iters)
```

```python
%matplotlib widget
plt.rcParams['axes.grid'] = False
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0], X[:,1], X[:,2], marker='o', c=idx, s=10, cmap='hsv')
plt.show()
```

```python
def feature_normalize(X):
    mu = np.mean(X, axis=0)
    sigma = np.std(X, ddof=1, axis=0)
    X_norm = (X - mu)/sigma
    
    return X_norm, mu, sigma
```

```python
def pca(X):
    # computing the covariance matrix
    sigma = (X.T @ X)/X.shape[0]
    U, S, X = np.linalg.svd(sigma)
    return U, S
```

```python
def project_data(X, U, K):
    U_reduce = U[:,:K]
    return X @ U_reduce 

def recover_data(Z, U, K):
    return Z @ U[:,:K].T
```

```python
X_norm, _, _ = feature_normalize(X)
U, _ = pca(X_norm)
Z = project_data(X_norm, U, 2)
```

```python
fig = plt.figure()
plt.scatter(Z[:,0], Z[:,1], c=idx, marker='*', cmap='hsv')
plt.show()
```
