```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from mpl_toolkits.mplot3d import Axes3D
from jupyterthemes import jtplot
jtplot.style()
```

```python
data_path = "ex1data1.txt"
with open(data_path, "r", newline="") as file:
    reader = csv.reader(file)
    data = np.array(list(reader), dtype=float)
```

```python
X = np.reshape(data[:,0], (-1,1))
y = np.reshape(data[:,1], (-1,1))
m = y.shape[0]
X = np.append(np.ones([m,1]), X, axis=1)
print(X.shape)
print(y.shape)
```

```python
a = data[:,0]
b = data[:,1]
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
plt.scatter(a, b, marker = 'x')
plt.show()
```

```python
def h_x(theta, X):
    return X.dot(theta)

def loss(h, y):
    m = y.shape[0]
    return np.sum((h-y)**2)/(2*m)
    
theta = np.zeros((2, 1))
h = h_x(theta, X)
loss(h, y)
```

```python
def gradient_descent(X, y, theta, alpha, num_iters):
    m = y.shape[0]
    loss_history = []
    for i in range(num_iters):
        temp0 = theta[0] - (alpha/m)*np.sum((h_x(theta, X) - y)) #*X[:,0]
        temp1 = theta[1] - (alpha/m)*np.sum((h_x(theta, X) - y)[:,0]*X[:,1])
        theta[0] = temp0
        theta[1] = temp1
        loss_history.append(loss(h_x(theta, X), y))
    
    return [theta, loss_history]

alpha = 0.01
N_of_iter = 2000
theta, loss_history = gradient_descent(X, y, theta, alpha, N_of_iter)
theta
```

```python
plt.plot(range(N_of_iter), loss_history)
plt.xlabel('Num of iteration');
plt.ylabel('Loss')
plt.show()
```

```python
a = data[:,0]
b = data[:,1]
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
plt.scatter(a, b, marker = 'x')
plt.plot(X[:,1], X.dot(theta))
plt.show()
```

```python
theta0_vals = np.linspace(-10, 10, 100)
theta1_vals = np.linspace(-1, 4, 100)

J_vals = np.zeros((theta0_vals.shape[0], theta0_vals.shape[0]))

for i in range(theta0_vals.shape[0]):
    for j in range(theta1_vals.shape[0]):
        temp_theta = np.vstack((theta0_vals[i], theta1_vals[j])) 
        J_vals[i,j] = loss(h_x(temp_theta, X), y)

J_vals.shape
```

```python
fig = plt.figure()
ax = Axes3D(fig)
#ax = fig.gca(projection='3d')

X_points, Y_points = np.meshgrid(theta0_vals, theta1_vals)

ax.plot_surface(Y_points, X_points, J_vals)
ax.set_xlabel('theta1')
ax.set_ylabel('theta0')
ax.set_zlabel('Loss')
ax.set_title('Surface plot')
plt.show()
```
