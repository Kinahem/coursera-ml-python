```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
from scipy.optimize import minimize
from jupyterthemes import jtplot
jtplot.style()
```

# Data processing

```python
data_path = "ex5data1.mat"
data = sio.loadmat(data_path)
print(data.keys())
```

```python
X = data['X']
y = data['y']
X_val = data['Xval']
y_val = data['yval']
X_test = data['Xtest']
y_test = data['ytest']
m = y.shape[0]
print("X shape : ", X.shape)
print("y shape : ", y.shape)
print("X_val shape : ", X_val.shape)
print("X_test shape : ", X_test.shape)
```

```python
plt.ylabel('Water flowing out of the dam (y)')
plt.xlabel('Change in water level (x)')
plt.scatter(X, y, marker='x', c='r')
plt.show()
```

## Functions

```python
def lin_cost(theta, X, y, lamb):
    m = X.shape[0]
    X = np.hstack((np.ones([m, 1]), X))
    theta = theta[None, :]
    h_x = X @ theta.T

    J = 1/(2*m)*np.sum((h_x - y)**2)
    J_reg = lamb/(2*m)*np.sum(theta[:, 1:]**2)
    J += J_reg

    grad = 1/m * (h_x - y).T @ X
    grad[:, 1:] += lamb/m * theta[:, 1:]

    return J, grad.ravel()


theta = np.array([1, 1])
J, grad = lin_cost(theta, X, y, 1)
print("J expected = 303.993, actual:", J)
print("grad expected = [-15.30, 598.250], actual:", grad)
```

```python
def train(arguments, disp=False):
    options = {'disp': disp}
    theta_size = arguments[0].shape[1] + 1
    initial_theta = np.zeros(theta_size)
    res = minimize(lin_cost, initial_theta,
                   jac=True, args=arguments,
                   method='TNC',  # BFGS or TNC
                   options=options)
    return res.x
```

# Train linear regression

```python
lamb = 1
theta = train((X, y, lamb), True)
theta
```

```python
plt.ylabel('Water flowing out of the dam (y)')
plt.xlabel('Change in water level (x)')
plt.scatter(X, y, marker='x', c='r')
plt.plot(X, np.hstack((np.ones([m, 1]), X))@theta[None, :].T)  # X @ theta
plt.show()
```

## Learning curve for linear regression

```python
def learning_curve(X, y, X_val, y_val, lamb):
    train_loss = []
    val_loss = []
    for i in range(1, m+1):
        theta = train((X[:i], y[:i], lamb))
        train_loss.append(lin_cost(theta, X[:i], y[:i], lamb)[0])
        val_loss.append(lin_cost(theta, X_val, y_val, lamb)[0])

    return train_loss, val_loss
```

```python
train_loss, val_loss = learning_curve(X, y, X_val, y_val, 1)
plt.ylabel('Error')
plt.xlabel('Num of training examples')
plt.title('Learning curve for linear regression')
plt.plot(range(1, m+1), train_loss)
plt.plot(range(1, m+1), val_loss)
plt.legend(["Train", "Validation"])
plt.show()
```

# Train polynomial regression with lambda = 0

```python
def poly_feat(X, n):
    for i in range(2, n+1):
        X = np.r_['1', X, (X[:, 0]**i)[:, None]]
    return X
```

```python
def feat_norm(X):
    mu = X.mean(axis=0)
    sigma = X.std(axis=0)
    X = (X - mu)/sigma
    return X, mu, sigma
```

```python
p = 8
X_poly = poly_feat(X, p)
X_poly, mu, sigma = feat_norm(X_poly)

X_poly_val = poly_feat(X_val, p)
X_poly_val = (X_poly_val - mu)/sigma

X_poly_test = poly_feat(X_test, p)
X_poly_test = (X_poly_test - mu)/sigma
```

```python
def plot_fit(poly_feat, min_X, max_X, mu, sigma, theta, p):
    X = np.linspace(min_X - 15, max_X + 25, 200)[:, None]
    m = X.shape[0]
    X_poly = poly_feat(X, p)
    X_poly = (X_poly - mu)/sigma
    X_poly = np.hstack((np.ones((m, 1)), X_poly))

    plt.plot(X, (X_poly@theta), '--', lw=2)
```

```python
plt.ylabel('Water flowing out of the dam (y)')
plt.xlabel('Change in water level (x)')
plt.title('Polynomial Regression Fit, lambda = 0')
theta = train((X_poly, y, 0))
plt.ylim((-5, 30))
plt.scatter(X, y, marker='x', c='r')
plot_fit(poly_feat, np.min(X), np.max(X), mu, sigma, theta, p)
plt.show()
```

```python
plt.ylabel('Water flowing out of the dam (y)')
plt.xlabel('Change in water level (x)')
plt.title('Polynomial Regression Fit on validation, lambda = 0')
plt.ylim((-5, 30))
plt.scatter(X_val, y_val, marker='x', c='r')
plot_fit(poly_feat, np.min(X_val), np.max(X_val), mu, sigma, theta, p)
plt.show()
```

## Learning curve for polynomial regression with lambda = 0

```python
train_loss, val_loss = learning_curve(X_poly, y, X_poly_val, y_val, 0)
plt.ylabel('Error')
plt.xlabel('Num of training examples')
plt.title('Polynomial Regression Learning Curve, lambda = 0')
plt.plot(range(1, m+1), train_loss)
plt.plot(range(1, m+1), val_loss)
plt.legend(["Train", "Validation"])
plt.show()
```

```python
from ipywidgets import interact
@interact
def lambda_train(lamb=(0, 10, 0.5)):
    theta = train((X_poly, y, lamb))
    print(lin_cost(theta, X_poly, y, lamb)[0])
    print(lin_cost(theta, X_poly_val, y_val, lamb)[0])
```

# Train polynomial regression with lambda = 100

```python
plt.ylabel('Water flowing out of the dam (y)')
plt.xlabel('Change in water level (x)')
plt.title('Polynomial Regression Fit, lambda = 100')
theta = train((X_poly, y, 100))
plt.ylim((-5, 30))
plt.scatter(X, y, marker='x', c='r')
plot_fit(poly_feat, np.min(X), np.max(X), mu, sigma, theta, p)
plt.show()
```

## Learning curve for polynomial regression with lambda = 100

```python
train_loss, val_loss = learning_curve(X_poly, y, X_poly_val, y_val, 100)
plt.ylabel('Error')
plt.xlabel('Num of training examples')
plt.title('Polynomial Regression Learning Curve, lambda = 100')
plt.plot(range(1, m+1), train_loss)
plt.plot(range(1, m+1), val_loss)
plt.legend(["Train", "Validation"])
plt.show()
```

## Validation curve for polynomial regression

```python
def validation_curve(X, y, X_val, y_val, lamb_vec):
    train_loss = []
    val_loss = []
    for lamb in lamb_vec:
        theta = train((X, y, lamb))
        train_loss.append(lin_cost(theta, X, y, lamb)[0])
        val_loss.append(lin_cost(theta, X_val, y_val, lamb)[0])

    return train_loss, val_loss
```

```python
lamb_vec = [0, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10]
train_loss, val_loss = validation_curve(X_poly, y, X_poly_val, y_val, lamb_vec)
plt.ylabel('Error')
plt.xlabel('Lambda')
plt.title('Polynomial Regression Learning Curve, lambda = 0')
plt.plot(lamb_vec, train_loss)
plt.plot(lamb_vec, val_loss)
plt.legend(["Train", "Validation"])
plt.show()
```

```python
np.set_printoptions(precision=5)
print("Lambda\tTrain Loss\tVal Loss")
for i, lamb in enumerate(lamb_vec):
    print("{:.3f}\t{:.6f}\t{:.6f}".format(lamb, train_loss[i], val_loss[i]))
```
